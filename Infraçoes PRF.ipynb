<h3>A)Obs: só foram computadas infrações no ano de 2021, portanto, a resposta é <strong>5241299</strong></h3>

<h3>B) A quantidade de variáveis pode ser obtida atraves do numero de colunas, para isso basta usar o len</h3>

<h3>C) Para obter as cidades com mais infraçoes, devemos contar a quantidade de ocorrencia dessas cidades em todo o dataframe, usando um value_counts. Para limitar em 5 ocorrencias, é ideal usar o head com o valor 5.</h3>

```python
#Quantas infrações ocorreram em cada ano?
quant_infracoes = len(dados["Data da Infração (DD/MM/AAAA)"])
print(f"infraçoes em 2021: {quant_infracoes}")


#Quantas variáveis cada conjunto de dados registra?
quant_variaveis = len(dados.columns)
print(f"Quantidade de variáveis: {quant_variaveis}")


#Top 5 cidades com mais infraçoes
cidades_mais_infratoras = dados["Município"].value_counts(sort=True)
pd.DataFrame(cidades_mais_infratoras)
cidades_mais_infratoras = cidades_mais_infratoras.reset_index()

cidades_mais_infratoras.rename(columns={"index":"Município", "Município":"Infrações"}, inplace=True)
cidades_mais_infratoras.head(5)
```

<h1><strong>Tarefa 3</strong></h1>

<h3>A)Para ranquear os estados levando em consideraçao os estados com mais infrações basta contar o <i>.value_counts</i> de "UF_INFRAÇÃO" e ordena-los usando o sort.</h3>

```python
ranking_estados = dados["UF Infração"].value_counts(sort=True)
ranking_estados = ranking_estados.reset_index()
ranking_estados.rename(columns={"index":"UF", "UF Infração": "Infrações"}, inplace=True)
print(pd.DataFrame(ranking_estados))


ranking_estados.to_csv("Ranking_estados.csv")


#por dia

#dados["Data da Infração (DD/MM/AAAA)"] = pd.to_datetime(dados["Data da Infração (DD/MM/AAAA)"])
#dias_dados = dados["Data da Infração (DD/MM/AAAA)"]
#pd.DataFrame(dias_dados)
#dias_dados.strftime("%A")

#Não consegui :(

```
